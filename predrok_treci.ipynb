```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from numpy import sign
from scipy.interpolate import lagrange
```

```python
def f(x):
  return x*np.exp(x**2)

a,b=2,3
Itrue=4024.242889
```

```python
def simpson_comp(f, a, b, n):
    if n % 2 == 1:
        raise ValueError("n mora biti paran")
    h = (b - a) / n
    x = np.linspace(a, b, n+1)
    y = f(x)
    I = h/3 * (y[0] + y[-1]
               + 4*np.sum(y[1:-1:2])
               + 2*np.sum(y[2:-2:2]))
    return I
```

```python
n=20
I_n = simpson_comp(f, a, b, n)
print("Simpson n=20:", I_n, "  greska:", abs(I_n - Itrue), f"relativna greska:{(abs(I_n - Itrue) / Itrue) * 100:.4f}%")

```

```python
eps=0.5e-3
I_2n = simpson_comp(f, a, b, n*2)
while(abs(I_2n-I_n)/15>eps):
  n*=2
  I_2n = simpson_comp(f, a, b, n)
  print("Simpson n=",n,":", I_2n, "  E=:", abs(I_2n - Itrue), f"relativna greska:{(abs(I_n - Itrue) / Itrue) * 100:.4f}%")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def f(x):
  return x**3 - 4*x + 1

a,b=0,2
Itrue=-2
```

1. Numerički izračunajte integral

• trapeznim pravilom,

• Simpsonovim pravilom.


Trapez

```python
def trapez_comp(f, a, b, n):
	h = (b - a) / n
	x = np.linspace(a, b, n+1)
	y = f(x)
	return h * (0.5*y[0] + np.sum(y[1:-1]) + 0.5*y[-1])

I_2 = trapez_comp(f, a, b, 2)
print("Trapez n=2:", I_2, "  greska:", abs(I_2 - Itrue), f"relativna greska:{(abs(I_2 - Itrue) / Itrue) * 100:.4f}%")

I_4 = trapez_comp(f, a, b, 4)
print("Trapez n=4:", I_4, "  greska:", abs(I_4 - Itrue), f"relativna greska:{(abs(I_4 - Itrue) / Itrue) * 100:.4f}%")

I_8 = trapez_comp(f, a, b, 8)
print("Trapez n=8:", I_8, "  greska:", abs(I_8 - Itrue), f"relativna greska:{(abs(I_8 - Itrue) / Itrue) * 100:.4f}%")

I_16 = trapez_comp(f, a, b, 16)
print("Trapez n=16:", I_16, "  greska:", abs(I_16 - Itrue), f"relativna greska:{(abs(I_16 - Itrue) / Itrue) * 100:.4f}%")

I_32 = trapez_comp(f, a, b, 32)
print("Trapez n=32:", I_32, "  greska:", abs(I_32 - Itrue), f"relativna greska:{(abs(I_32 - Itrue) / Itrue) * 100:.4f}%")

I_64 = trapez_comp(f, a, b, 64)
print("Trapez n=64:", I_64, "  greska:", abs(I_64 - Itrue), f"relativna greska:{(abs(I_64 - Itrue) / Itrue) * 100:.4f}%")

I_128 = trapez_comp(f, a, b, 128)
print("Trapez n=128:", I_128, "  greska:", abs(I_128 - Itrue), f"relativna greska:{(abs(I_128 - Itrue) / Itrue) * 100:.4f}%")

I_500 = trapez_comp(f, a, b, 500)
print("Trapez n=500:", I_500, "  greska:", abs(I_500 - Itrue), f"relativna greska:{(abs(I_500 - Itrue) / Itrue) * 100:.4f}%")
```

Simpson

```python
def simpson_comp(f, a, b, n):
    if n % 2 == 1:
        raise ValueError("n mora biti paran")
    h = (b - a) / n
    x = np.linspace(a, b, n+1)
    y = f(x)
    I = h/3 * (y[0] + y[-1]
               + 4*np.sum(y[1:-1:2])
               + 2*np.sum(y[2:-2:2]))
    return I

IS_2 = simpson_comp(f, a, b, 2)
print("Simpson n=2:", IS_2, "  greska:", abs(IS_2 - Itrue), f"relativna greska:{(abs(IS_2 - Itrue) / Itrue) * 100:.4f}%")

IS_4= simpson_comp(f, a, b, 4)
print("Simpson n=4:", IS_4, "  greska:", abs(IS_4 - Itrue), f"relativna greska:{(abs(IS_4 - Itrue) / Itrue) * 100:.4f}%")

IS_8 = simpson_comp(f, a, b, 8)
print("Simpson n=8:", IS_8, "  greska:", abs(IS_8 - Itrue), f"relativna greska:{(abs(IS_8 - Itrue) / Itrue) * 100:.4f}%")

IS_16 = simpson_comp(f, a, b, 16)
print("Simpson n=16:", IS_16, "  greska:", abs(IS_16 - Itrue), f"relativna greska:{(abs(IS_16 - Itrue) / Itrue) * 100:.4f}%")
```

3. Aproksimirajte derivaciju funkcije g(x) = ln x u točki x = 1:

• diferencijom unaprijed,

• centralnom diferencijom.

```python
def g(x):
  return np.log(x)

def forward(x):
  return (g(x+h) - g(x)) / h

def central(x):
  return (g(x+h) - g(x-h)) / (2*h)

h = 0.05
print("Derivacija u x=1 diferencijom unaprijed:",forward(1))
print("Derivacija u x=1 centralnom diferencijom:",central(1))
```

**točna vrijednost bi trebala biti 1, tako da je centralna diferencija puno preciznija, iako i dalje ima odstupanje**

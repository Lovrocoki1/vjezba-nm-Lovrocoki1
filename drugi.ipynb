```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from numpy import sign
from scipy.linalg import lu, lu_factor, lu_solve
```

```python
# Definicija matrice A i vektora b
A = np.array([[3,-0.1,-0.2], [0.1,7,-0.3],[0.3,-0.2,10]])
b = np.array([7.85,-19.3,71.4])
```

Stvarna rješenja su x1=3, x2=-2.5, x3=7

1. Riješite sustav:

• Gaussovom eliminacijom s parcijalnim pivotiranjem,

• LU dekompozicijom.


Gauss

```python
def gauss_solve(A, b):
    A = A.astype(float).copy()
    b = b.astype(float).copy()
    n = len(b)

    # 1) Forward eliminacija s djelomičnim pivotiranjem
    for k in range(n-1):
        # odaberi pivot red (najveći po apsolutnoj vrijednosti u stupcu k)
        pivot = k + np.argmax(np.abs(A[k:, k]))
        if A[pivot, k] == 0:
            raise ValueError("Matrica je singularna ili blizu singularne.")
        # zamijeni redove u A i b ako treba
        if pivot != k:
            A[[k, pivot]] = A[[pivot, k]]
            b[[k, pivot]] = b[[pivot, k]]

        # eliminiraj ispod pivota
        for i in range(k+1, n):
            lam = A[i, k] / A[k, k]
            A[i, k:] = A[i, k:] - lam * A[k, k:]
            b[i] = b[i] - lam * b[k]

    # 2) Povratna (unatrag) supstitucija
    x = np.zeros(n)
    for i in range(n-1, -1, -1):
        s = np.dot(A[i, i+1:], x[i+1:])
        x[i] = (b[i] - s) / A[i, i]
    return x

x_gauss = gauss_solve(A, b)
print("\nRješenje sustava:")
print("x1 =", x_gauss[0])
print("x2 =", x_gauss[1])
print("x3 =", x_gauss[2])
```

LU

```python
# Izračun LU dekompozicije sa permutacijom
P, L, U = lu(A)

print("Matrica P (permutacijska):")
print(P)
print("\nMatrica L (donje trokutasta):")
print(L)
print("\nMatrica U (gornje trokutasta):")
print(U)

# Brzo rješenje koristeći LU faktoriranje
lu_and_piv = lu_factor(A)
#x = lu_solve(lu_and_piv, b)
x_LU=np.linalg.solve(A, b)
print("\nRješenje sustava:")
print("x1 =", x_LU[0])
print("x2 =", x_LU[1])
print("x3 =", x_LU[2])
```

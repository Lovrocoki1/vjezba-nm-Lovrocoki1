**4. zadatak - Aproksimacija i metoda najmanjih kvadrata (LV4)**




```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

Zadani su podatci mjerenja: (1, 1.1), (2, 1.9), (3, 3.2), (4, 3.9), (5, 5.1)

```python
x = np.array([1, 2, 3, 4, 5])
A = np.column_stack((np.ones(len(x)), x))
b=np.array([1.1,1.9,3.2,3.9,5.1])
```

1. Odredite aproksimacijski pravac y = a0 + a1x metodom najmanjih kvadrata:

• pomocu QR dekompozicije,

• pomocu SVD metode.

QR

```python
#EkonomicnaQRdekompozicija
Q,R=np.linalg.qr(A,mode="reduced")
#Korak2:c=Q^Tb
c=Q.T@b
#Korak3:rjesenjeRx=c(backsubstitution)
x=np.linalg.solve(R,c)
print("Q =\n",Q)
print("R =\n",R)
print("Rjesenjenajmanjihkvadratax* =",x)
```

SVD

```python
#SVDdekompozicija
U,s,Vt =np.linalg.svd(A,full_matrices=False)
Sigma_plus=np.diag(1.0/s)
#y=U^Tb
y=U.T@b
#x* =VSigma^+y
x=Vt.T @(Sigma_plus@y)
print("Singularne vrijednosti:",s)
print("Rjesenje najmanjih kvadrata x* =",x)
```

2. Izračunajte normu reziduala.


```python

#QRrjesenje
Q,R= np.linalg.qr(A)
a_qr= np.linalg.solve(R,Q.T@b)
#SVDrjesenje
U, S, Vt = np.linalg.svd(A, full_matrices=False)
a_svd = Vt.T @ (np.diag(1/S) @ (U.T @ b))
# Reziduali
res_qr = np.linalg.norm(A @ a_qr-b)
res_svd = np.linalg.norm(A @ a_svd- b)
print(a_qr, a_svd)
print(res_qr, res_svd)
```
